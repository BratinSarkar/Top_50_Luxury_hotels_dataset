```python
import pandas as pd
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
# Step 1: Combine relevant columns into a single feature
df['Features'] = df['Overview'] + " " + df['Dining_Area'] + " " + df['Drinking_Area'] + " " + df['Hotel_Ammenties']

# Step 2: Transform text data into TF-IDF features
vectorizer = TfidfVectorizer(stop_words='english')
tfidf_matrix = vectorizer.fit_transform(df['Features'])

# Step 3: Compute cosine similarity between hotels
cosine_sim = cosine_similarity(tfidf_matrix, tfidf_matrix)

# Step 4: Define recommendation function
def recommend_hotel(hotel_name, cosine_sim=cosine_sim, df=df):
    if hotel_name not in df['Name'].values:
        return f"Hotel '{hotel_name}' not found in the dataset."

    # Find the index of the hotel
    idx = df[df['Name'] == hotel_name].index[0]

    # Get similarity scores and sort them
    sim_scores = list(enumerate(cosine_sim[idx]))
    sim_scores = sorted(sim_scores, key=lambda x: x[1], reverse=True)

    # Get top 5 similar hotels (excluding the queried hotel itself)
    sim_scores = sim_scores[1:6]

    # Get hotel names
    recommended_hotels = [df['Name'].iloc[i[0]] for i in sim_scores]

    return recommended_hotels

# Step 5: Test the recommendation system
hotel_to_recommend = 'Capella Bangkok'
recommendations = recommend_hotel(hotel_to_recommend)
print(f"Hotels similar to '{hotel_to_recommend}':")
for rec in recommendations:
    print(f"- {rec}")
```

```python
# Define traveler personas
def identify_persona(amenities):
    if 'family friendly' in amenities:
        return 'Family Traveler'
    elif 'spa' in amenities:
        return 'Wellness Enthusiast'
    elif 'beach access' in amenities:
        return 'Beach Lover'
    else:
        return 'General Luxury Traveler'

# Apply persona identification
df['Traveler Persona'] = df['Hotel_Ammenties'].apply(identify_persona)

# Count by personas
persona_counts = df['Traveler Persona'].value_counts()

# Plot personas
plt.figure(figsize=(8, 5))
sns.barplot(x=persona_counts.index, y=persona_counts.values, palette='Set2')
plt.title('Luxury Traveler Personas', fontsize=16)
plt.xlabel('Persona')
plt.ylabel('Count of Hotels')
plt.xticks(rotation=45, ha='right')
plt.show()
```

```python
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.cluster import KMeans
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# ... (Your previous code for data loading and preparation) ...

# Vectorize the Hotel Amenities
vectorizer = CountVectorizer(stop_words='english')
amenities_matrix = vectorizer.fit_transform(df['Hotel_Ammenties'])

# Perform K-Means Clustering
kmeans = KMeans(n_clusters=3, random_state=42)  # Adjust clusters as needed
df['cluster'] = kmeans.fit_predict(amenities_matrix)

# Analyze Clusters (print cluster characteristics)
print(df.groupby('cluster')['Hotel_Ammenties'].apply(lambda x: ' | '.join(x[:5])))

# Visualize Clusters (using PCA for dimensionality reduction)
from sklearn.decomposition import PCA

pca = PCA(n_components=2)  # Reduce to 2 dimensions for plotting
principal_components = pca.fit_transform(amenities_matrix.toarray())
principal_df = pd.DataFrame(data=principal_components, columns=['PC1', 'PC2'])
principal_df['cluster'] = df['cluster']

plt.figure(figsize=(8, 6))
sns.scatterplot(x='PC1', y='PC2', hue='cluster', data=principal_df, palette='viridis', s=100)
plt.title('Visualization of Hotel Clusters based on Amenities')
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.show()
```
